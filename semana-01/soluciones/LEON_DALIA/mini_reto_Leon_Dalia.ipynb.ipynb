```python
# Crear archivo Java
codigo_java = '''
import java.util.Scanner;

public class LoginSeguro {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        final int MAX_INTENTOS = 3;
        int intentos = 0;
        boolean accesoConcedido = false;

        System.out.println("=== SISTEMA DE LOGIN SEGURO ===");
        System.out.println("Usuario de prueba: admin");
        System.out.println("Contraseña de prueba: Admin123!");
        System.out.println("================================\\n");

        while (intentos < MAX_INTENTOS && !accesoConcedido) {
            System.out.println("--- Intento " + (intentos + 1) + " de " + MAX_INTENTOS + " ---");

            System.out.print("Usuario: ");
            String usuario = sc.nextLine().trim();

            System.out.print("Contraseña: ");
            String password = sc.nextLine();

            // Validaciones
            boolean esValido = true;

            // Validar longitud mínima 8
            if (password.length() < 8) {
                System.out.println("[ERROR] La contraseña debe tener al menos 8 caracteres");
                esValido = false;
            }

            // Validar al menos un número
            if (!password.matches(".*[0-9].*")) {
                System.out.println("[ERROR] La contraseña debe contener al menos un número (0-9)");
                esValido = false;
            }

            // Validar al menos una mayúscula
            if (!password.matches(".*[A-Z].*")) {
                System.out.println("[ERROR] La contraseña debe contener al menos una mayúscula (A-Z)");
                esValido = false;
            }

            // Verificar credenciales
            if (esValido && usuario.equals("admin") && password.equals("Admin123!")) {
                System.out.println("\\n✓ ACCESO CONCEDIDO. Bienvenido, " + usuario);
                accesoConcedido = true;
            } else if (esValido) {
                System.out.println("✗ Credenciales inválidas\\n");
                intentos++;
            } else {
                System.out.println("✗ Validaciones fallidas - Intento inválido\\n");
                intentos++;
            }

            if (intentos == MAX_INTENTOS && !accesoConcedido) {
                System.out.println("\\n⚠ CUENTA BLOQUEADA TEMPORALMENTE");
                System.out.println("Demasiados intentos fallidos (" + MAX_INTENTOS + ")");
            }
        }

        sc.close();
    }
}
'''

# Guardar archivo
with open('LoginSeguro.java', 'w') as f:
    f.write(codigo_java)

print("✅ Archivo LoginSeguro.java creado")
```

```python
# Código para ejecutar Java interactivamente en Colab
import subprocess
import sys

def ejecutar_java_interactivo(codigo_java):
    with open('Temp.java', 'w') as f:
        f.write(codigo_java)

    # Compilar
    subprocess.run(['javac', 'Temp.java'], capture_output=True)

    # Ejecutar
    process = subprocess.Popen(['java', 'Temp'],
                               stdin=subprocess.PIPE,
                               stdout=subprocess.PIPE,
                               stderr=subprocess.PIPE,
                               text=True)

    print("=== EJECUTANDO ===")
    print("(Escribe tus datos. Presiona Enter después de cada línea)")
    print("Para terminar, presiona Ctrl+C\\n")

    # Aquí puedes interactuar
    stdout, stderr = process.communicate(input=input("Ingresa usuario y contraseña: "))
    print(stdout)

# Versión simplificada
codigo_simple = '''
import java.util.Scanner;
public class Temp {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        System.out.print("Usuario: ");
        String user = sc.nextLine();
        System.out.print("Contraseña: ");
        String pass = sc.nextLine();

        if(pass.length() >= 8 && pass.matches(".*[0-9].*") && pass.matches(".*[A-Z].*")) {
            System.out.println("✓ Acceso concedido");
        } else {
            System.out.println("✗ Contraseña no cumple requisitos");
        }
        sc.close();
    }
}
'''
```
